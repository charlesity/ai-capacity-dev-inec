| **Fill with mean** | For numeric columns with normal distribution |
| **Fill with median** | For numeric columns with outliers |
| **Fill with mode** | For categorical columns |
| **Fill with constant** | When you know the default value |

# 📊 SECTION 7: HANDLING OUTLIERS

## Detecting and Handling Outliers

### 📖 What You Need to Know

Outliers are extreme values that can skew your analysis. The **IQR (Interquartile Range)** method is a common technique:

1. Calculate Q1 (25th percentile) and Q3 (75th percentile)
2. Calculate IQR = Q3 - Q1
3. Define bounds:
   - Lower bound = Q1 - 1.5 × IQR
   - Upper bound = Q3 + 1.5 × IQR
4. Identify outliers as values outside these bounds

### ✏️ Your Task

Detect and handle outliers in the `Total_Votes_Cast` column.
