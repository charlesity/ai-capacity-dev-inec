```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================================
# Original class distribution (encoded labels)
# ==========================================
# 0, 1, 2, 3 correspond to your encoded classes
original = {
    0: 1099248,    # Class 0
    1: 13029,      # Class 1
    2: 33965405,   # Class 2 (Majority)
    3: 3301524     # Class 3
}

# ==========================================
# After SMOTE
# ==========================================
majority = max(original.values())

smote = {
    0: majority,
    1: majority,
    2: majority,
    3: majority
}

classes = list(original.keys())

orig_counts = [original[c] for c in classes]
smote_counts = [smote[c] for c in classes]

x = np.arange(len(classes))
width = 0.35

plt.figure(figsize=(8,5))

plt.bar(
    x - width/2,
    orig_counts,
    width,
    label='Original',
    color='steelblue',
    edgecolor='black'
)

plt.bar(
    x + width/2,
    smote_counts,
    width,
    label='After SMOTE',
    color='darkorange',
    edgecolor='black'
)

plt.xticks(x, ['0', '1', '2', '3'], fontsize=12)
plt.xlabel('Themal Risk Class', fontsize=12)
plt.ylabel('Number of Samples', fontsize=12)
plt.title('', fontsize=14)
plt.legend(fontsize=11, bbox_to_anchor=(0.5, -0.2), loc='upper center')
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```
